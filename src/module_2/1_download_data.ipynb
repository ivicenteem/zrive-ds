```python
import os
import boto3
from dotenv import load_dotenv

load_dotenv('../../.env')

s3 = boto3.client(
    's3',
    aws_access_key_id=os.getenv('AWS_ACCESS_KEY_ID'),
    aws_secret_access_key=os.getenv('AWS_SECRET_ACCESS_KEY')
)

bucket = 'zrive-ds-data'
local_dir = '../../data/raw'
os.makedirs(local_dir, exist_ok=True)

# Rutas REALES encontradas en el servidor
files = [
    'groceries/sampled-datasets/orders.parquet',
    'groceries/sampled-datasets/regulars.parquet',
    'groceries/sampled-datasets/abandoned_carts.parquet',
    'groceries/sampled-datasets/inventory.parquet',
    'groceries/sampled-datasets/users.parquet',
    'groceries/box_builder_dataset/feature_frame.csv'    
]

for f in files:
    name = f.split('/')[-1]
    path = os.path.join(local_dir, name)
    
    if not os.path.exists(path):
        print(f"Descargando: {name}...")
        s3.download_file(bucket, f, path)
    else:
        print(f"OK: {name} ya existe.")

print("\nTodos los archivos listos en data/raw/")
```
